```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec

# --- Load Data ---
# fig5b data
Jx = np.load("fig5data/Jlistre0.35.npy")
Jy = np.load("fig5data/Jlistim0.35.npy")
reJCD12 = np.load("fig5data/JCD1_2_J30pt5Jminpt01_deltapt35MHz_g1pt85_T200ns.npy")
reJCD21 = np.load("fig5data/JCD2_1_J30pt5Jminpt01_deltapt35MHz_g1pt85_T200ns.npy")
ImJCDdiff = np.load("fig5data/ImJCDdiff_J30pt5Jminpt01_deltapt35MHz_g1pt85_T200ns.npy")
T_loop = 0.2
tt = np.linspace(0, T_loop, int(T_loop * 1000))

# fig5c data
bloch_inst = np.load("fig5data/bloch_instantaneous_Delta0.35.npy")
bloch_nh = np.load("fig5data/bloch_measured_Delta0.35.npy")
trdist1 = np.load('fig5data/trdist1.npy')
times = np.linspace(0, 0.2, bloch_nh.shape[1])

# fig5d data
TrD_pt35 = np.load("fig5data/trace_distance_Delta0.35.npy")

# --- Layout: 2 rows, 3 columns
fig = plt.figure(figsize=(12, 4))
gs = gridspec.GridSpec(2, 3, height_ratios=[1, 2], width_ratios=[1, 1, 1], figure=fig)

# ---- fig5b 
ax_b_top = fig.add_subplot(gs[0, 0])
ax_b_top.plot(tt, Jy, label='$J_y$')
ax_b_top.plot(tt, ImJCDdiff, label='Im$[J_{CD}^{diff}]$')
ax_b_top.set_xlabel('Time $\mu$s')
ax_b_top.set_ylabel('Amplitude')
ax_b_top.set_title('(b)')
ax_b_top.legend()

# ---- fig5b
ax_b_bottom = fig.add_subplot(gs[1, 0])
ax_b_bottom.plot(tt, (reJCD12 - reJCD21) / 2, label='Re$[(J_{12}-J_{21})/2]$')
ax_b_bottom.plot(tt, ImJCDdiff, label='Im$[J_{CD}^{diff}]$')
ax_b_bottom.set_xlabel('Time $\mu$s')
ax_b_bottom.set_ylabel('Amplitude')
ax_b_bottom.legend()

# ---- fig5c 
ax_c = fig.add_subplot(gs[:, 1])
ax_c.plot(times, bloch_inst[2], 'k--')
ax_c.plot(times, bloch_inst[1], 'r--')
ax_c.plot(times, bloch_inst[0], 'b--')
ax_c.plot(times, bloch_nh[2], 'k', label='z')
ax_c.plot(times, bloch_nh[0], 'b', label='x')
ax_c.plot(times, bloch_nh[1], 'r', label='y')
ax_c.set_xlabel(r'Time $\mu$s')
ax_c.set_ylabel(r'$\langle\sigma\rangle$')
ax_c.set_title('(c)')
ax_c.legend()

# ---- fig5d 
ax_d = fig.add_subplot(gs[:, 2])
ax_d.plot(times, TrD_pt35, label='TrD')
ax_d.plot(times, trdist1, label='TrD inst.')
ax_d.set_ylim([0, 0.64])
ax_d.set_xlabel(r'Time $\mu$s')
ax_d.set_ylabel(r'$\overline{D}$')
ax_d.set_title('(d)')
ax_d.legend()

plt.tight_layout()
plt.show()
```
